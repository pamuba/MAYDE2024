```python
house_df_sample = house_df[['price', 'bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'sqft_above', 'sqft_basement', 'yr_built']]
```

```python
sns.pairplot(house_df_sample)
```

```python
selected_features = ['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long', 'sqft_living15', 'sqft_lot15']

X = house_df[selected_features]
```

```python
y = house_df['price']
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression(fit_intercept=True)
lm.fit(X_train,y_train)
print("Linear Model Coefficient (m): ", lm.coef_)
print("Linear Model Coefficient (b): ", lm.intercept_)
```

```python
y_predict = lm.predict(X_test)
y_predict
```

```python
plt.plot(y_test, y_predict, 'o', color='b')
plt.xlim(0,3000000)
plt.ylim(0,3000000)
plt.xlabel("Model Predictions")
plt.ylabel("True Vlaues")
plt.title("Linear Regression Predictions")
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt

k = X_test.shape[1]
n = len(X_test)

RMSE = float(format(sqrt(mean_squared_error(y_test, y_predict)),'.3f'))
MSE = mean_squared_error(y_test, y_predict)
MAE = mean_absolute_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)

print('RMSE=',RMSE,'\nMSE=',MSE,'\nMAE=',MAE,'\nR2=',r2, '\nAdjusted R2=',adj_r2)
```

```python
from sklearn.linear_model import Lasso, Ridge
regressor_ridge=Ridge(alpha=30)
regressor_ridge.fit(X_train, y_train)
print("Linear Model Coefficient (m): ", regressor_ridge.coef_)
print("Linear Model Coefficient (b): ", regressor_ridge.intercept_)

y_predict = regressor_ridge.predict(X_test)
y_predict
```
